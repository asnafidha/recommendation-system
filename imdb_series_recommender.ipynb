**1:Defining the recommender function**

```python
# Define a function to recommend series based on user preferences
def recommend_series(genre=None, min_rating=None, top_n=5):
    print("recommendations based on user preferences")

    recommendations = df.copy()

    if genre:
        recommendations = recommendations[recommendations['Genre'].str.contains(genre, case=False)]


    if min_rating:
        recommendations = recommendations[recommendations['IMDB_Rating'] >= min_rating]


    recommendations = recommendations.sort_values(by='IMDB_Rating', ascending=False)

    return recommendations.head(top_n)

```

**3:Example Usage and Printing Top Recommendations**

```python
genre = 'Fantasy'
min_rating = 8.0
print(f"Recommendations for genre '{genre}' with minimum rating {min_rating}:")
top_recommendations = recommend_series(genre=genre, min_rating=min_rating, top_n=10)

print("Top Recommendations:")
print(top_recommendations[['Series_Title', 'Genre', 'IMDB_Rating',]])
```

**3:Finding Similar Series to a Reference Title**

-It uses the genre of the reference series to recommend similar series.

```python
def find_similar_series(reference_title, min_rating=None, top_n=5):
    print(f"Finding series similar to '{reference_title}'")


    reference_series = df[df['Series_Title'] == reference_title]
    if reference_series.empty:
        print("Reference series not found.")
        return

    reference_genre = reference_series['Genre'].iloc[0]

    top_recommendations = recommend_series(genre=reference_genre, min_rating=min_rating, top_n=top_n)

    print(f"Top Recommendations similar to '{reference_title}':")
    print(top_recommendations[['Series_Title', 'Genre', 'IMDB_Rating',]])
```
