```python
def weighted_rating(df):

   #1
    m = df['vote_count'].quantile(0.8)
    print(f"Minimum votes required for consideration: {m}")

    #2
    qualified_movies = df[df['vote_count'] >= m]
    print(f"Number of qualified movies: {len(qualified_movies)}")

    #3
    qualified_movies = qualified_movies[(qualified_movies['runtime'] >= 45) & (qualified_movies['runtime'] <= 300)]
    print(f"Number of qualified movies within runtime range: {len(qualified_movies)}")


    qualified_movies['score'] = (qualified_movies['vote_count'] / (qualified_movies['vote_count'] + m)) * qualified_movies['vote_average'] + (m / (qualified_movies['vote_count'] + m)) * qualified_movies['vote_average'].mean()


    qualified_movies = qualified_movies.sort_values(by='score', ascending=False)
    print("Top 5 movies based on IMDb weighted rating:")
    print(qualified_movies[['title', 'vote_count', 'vote_average', 'score']].head(10))

    return qualified_movies

result = weighted_rating(malayalam_movies)
```
